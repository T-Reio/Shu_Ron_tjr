```python
import pandas as pd
import numpy as np
```

```python
season = list(range(1957,2018))
salary_season = list(range(1987,2018))
```

```python
stats = pd.read_csv('master_thesis/input/f_battingstats_since57.csv')
```

```python
stats = stats.rename(columns = {'Name':'NAME', 'Team':'TEAM_played'})
stats = stats.replace({
    'Giants':'SFG', 'Cardinals':'STL', 'Indians':'CLE', 'Rockies':'COL', 'Blue Jays':'TOR', 'Athletics':'OAK',
    'Cubs':'CHC', 'Nationals':'WSH', 'Red Sox':'BOS', 'White Sox':'CHW', 'Tigers':'DET', 'Twins':'MIN',
    'Diamondbacks':'ARI', 'Orioles':'BAL', 'Reds':'CIN', 'Astros':'HOU',  'Royals':'KCR', 'Angels':'LAA',
    'Dodgers':'LAD', 'Marlins':'MIA', 'Yankees':'NYY', 'Mets':'NYM', 'Phillies':'PHI', 'Pirates':'PIT',
    'Padres':'SDP', 'Mariners':'SEA', 'Rays':'TBR', 'Rangers':'TEX', 'Brewers':'MIL', 'Braves':'ATL',
    '- - -':'PTms', 'Devil Rays':'TBR', 'Expos':'WSH'
    })
```

```python
stats.loc[stats['AVG'] >= 0.3,'AVG_above'] = True
stats.loc[stats['AVG'] < 0.3,'AVG_above'] = False
print(stats.head(5))
```

```python
stats.to_csv('master_thesis/output/battingstats_since57.csv', encoding = 'utf-8')
```

```python
play_stats = {}
for x in salary_season:
    play_stats[x] = stats[stats['Season'] == x]
    
print(play_stats[1987].head(5))
```

```python
salary_UST = {}
for x in salary_season:
    salary_UST[x] = pd.read_excel('master_thesis/input/MLB_stats_{}.xlsx'.format(x), sheet_name = 'salary_UST')
```

```python
for x in salary_season:
    salary_UST[x]['Season'] = x
    salary_UST[x] = salary_UST[x].rename(columns = {'TEAM':'TEAM_nextyr'})

for x in salary_season:
    salary_UST[x]['TEAM_nextyr'] = salary_UST[x]['TEAM_nextyr'].replace({
    'SF':'SFG', 'CWS':'CHW', 'KC':'KCR', 'SD':'SDP', 'TB':'TBR'
    })
    
for x in salary_season:
    print(salary_UST[x].head(3))
```

```python
stats_salary = {}

for x in salary_season:
    stats_salary[x] = pd.merge(play_stats[x], salary_UST[x], on = ['NAME','Season'], how = 'left')
        
print(stats_salary[2002].head(5))
```

```python
stats_salary = pd.concat(stats_salary[x] for x in salary_season)
```

```python
stats_salary.loc[stats_salary['AVG'] >= 0.3,'AVG_above'] = True
stats_salary.loc[stats_salary['AVG'] < 0.3,'AVG_above'] = False
stats_salary['Log AVG ANNUAL'] = np.log(stats_salary['AVG ANNUAL'])
stats_salary.loc[stats_salary['Log AVG ANNUAL'] == -np.inf, 'Log AVG ANNUAL'] = 0
print(np.isinf(stats_salary['Log AVG ANNUAL']).any())
stats_salary.loc[stats_salary['Season'] <= 1994, 'Era'] = 'bfst'
stats_salary.loc[(stats_salary['Season'] >= 1996) & (stats_salary['Season'] <= 2001), 'Era'] = 'stmb'
stats_salary.loc[(stats_salary['Season'] >= 2002), 'Era'] = 'afmb'
stats_salary.loc[stats_salary['Season'] == 1995, 'Era'] == 'Striked'
print(stats_salary.head(5))
print(stats_salary.tail(5))
```

```python
stats_salary.to_csv('master_thesis/output/stats_salary_87to17.csv', encoding = 'utf-8')
```
