```python
import pandas as pd
import numpy as np
```

```python
season = list(range(1987,2018))
```

```python
play = pd.read_csv('master_thesis/input/fstats_5718.csv')

sal = {}
for x in season:
    sal[x] = pd.read_excel('master_thesis/input/MLB_stats_{}.xlsx'.format(x), sheet_name = 'salary_UST')
    
fa = {}
for x in season:
    fa[x] = pd.read_excel('master_thesis/input/MLB_service_{}.xlsx'.format(x), sheet_name = 'FA')
```

```python
play = play.rename(columns = {'Name':'NAME', 'Team':'TEAM_played'})
play = play.replace({
    'Giants':'SFG', 'Cardinals':'STL', 'Indians':'CLE', 'Rockies':'COL', 'Blue Jays':'TOR', 'Athletics':'OAK',
    'Cubs':'CHC', 'Nationals':'WSH', 'Red Sox':'BOS', 'White Sox':'CHW', 'Tigers':'DET', 'Twins':'MIN',
    'Diamondbacks':'ARI', 'Orioles':'BAL', 'Reds':'CIN', 'Astros':'HOU',  'Royals':'KCR', 'Angels':'LAA',
    'Dodgers':'LAD', 'Marlins':'MIA', 'Yankees':'NYY', 'Mets':'NYM', 'Phillies':'PHI', 'Pirates':'PIT',
    'Padres':'SDP', 'Mariners':'SEA', 'Rays':'TBR', 'Rangers':'TEX', 'Brewers':'MIL', 'Braves':'ATL',
    '- - -':'PTms', 'Devil Rays':'TBR', 'Expos':'WSH'
    })
stats = {}
for x in season:
    stats[x] = play[play['Season'] == x]

print(stats[1987].head(5))
```

```python
for x in season:
    sal[x]['Season'] = x
    sal[x] = sal[x].rename(columns = {'TEAM':'TEAM_nextyr'})

for x in season:
    sal[x]['TEAM_nextyr'] = sal[x]['TEAM_nextyr'].replace({
    'SF':'SFG', 'CWS':'CHW', 'KC':'KCR', 'SD':'SDP', 'TB':'TBR'
    })
    
print(sal[1987].head(3))
```

```python
fa_fil = ['Name', 'Date', 'WAR3', 'Yrs', 'Season', 'FA']
for x in season:
    fa[x]['Season'] = x
    fa[x]['FA'] = 1
    fa[x] = fa[x].filter(fa_fil)
    fa[x] = fa[x].rename(columns = {'Name':'NAME'})
    
print(fa[1987].head(5))
```

```python
stats_salary = {}

for x in season:
    stats_salary[x] = pd.merge(stats[x], sal[x], on = ['NAME','Season'], how = 'right')
    
print(stats_salary[1987].head(5))
```

```python
stats_salary_fa = {}

for x in season:
    stats_salary_fa[x] = pd.merge(stats_salary[x], fa[x], on = ['NAME', 'Season'], how = 'left')
    
for x in season:
    print(stats_salary_fa[x].head(3))
```

```python
for x in season:
    stats_salary_fa[x]['FA'] = stats_salary_fa[x]['FA'].fillna(0)
    
stats_salary_fa[1996].head(5)
```

```python
avg_sal = {}
for x in season:
    stats_salary_fa[x] = stats_salary_fa[x][stats_salary_fa[x]['AVG ANNUAL'] != 0]
    avg_sal[x] = np.log(np.mean(stats_salary_fa[x]['AVG ANNUAL']))
```

```python
for x in season:
    stats_salary_fa[x]['Log AVG ANNUAL'] = np.log(stats_salary_fa[x]['AVG ANNUAL'])
    stats_salary_fa[x]['Dev Log salary'] = stats_salary_fa[x]['Log AVG ANNUAL'] - avg_sal[x]
```

```python
stats_salary_fa = pd.concat(stats_salary_fa[x] for x in season)
```

```python
stats_salary_fa.loc[stats_salary_fa['AVG'] >= 0.3,'AVG_300'] = 1
stats_salary_fa.loc[stats_salary_fa['AVG'] < 0.3,'AVG_300'] = 0
stats_salary_fa.loc[stats_salary_fa['HR'] >= 20, 'HR_20'] = 1
stats_salary_fa.loc[stats_salary_fa['HR'] < 20, 'HR_20'] = 0
stats_salary_fa.loc[stats_salary_fa['SB'] >= 30, 'SB_30'] = 1
stats_salary_fa.loc[stats_salary_fa['SB'] < 30, 'SB_30'] = 0
stats_salary_fa.loc[stats_salary_fa['SB'] >= 40, 'SB_40'] = 1
stats_salary_fa.loc[stats_salary_fa['SB'] < 40, 'SB_40'] = 0
stats_salary_fa.loc[stats_salary_fa['RBI'] >= 100, 'RBI_100'] = 1
stats_salary_fa.loc[stats_salary_fa['RBI'] < 100, 'RBI_100'] = 0
stats_salary_fa.loc[stats_salary_fa['OBP'] >= 0.35,'OBP_350'] = 1
stats_salary_fa.loc[stats_salary_fa['OBP'] < 0.35,'OBP_350'] = 0
stats_salary_fa.loc[stats_salary_fa['OPS'] >= 1,'OPS_1000'] = 1
stats_salary_fa.loc[stats_salary_fa['OPS'] < 1,'OPS_1000'] = 0
stats_salary_fa.loc[stats_salary_fa['PA'] >= 500,'PA_500'] = 1
stats_salary_fa.loc[stats_salary_fa['PA'] < 500,'PA_500'] = 0
stats_salary_fa.loc[stats_salary_fa['H'] >= 200,'H_200'] = 1
stats_salary_fa.loc[stats_salary_fa['H'] < 200,'H_200'] = 0
```

```python
stats_salary_fa.loc[stats_salary_fa['Season'] <= 1994, 'Era'] = 'bfst'
stats_salary_fa.loc[(stats_salary_fa['Season'] >= 1995) & (stats_salary_fa['Season'] <= 2002), 'Era'] = 'stmb'
stats_salary_fa.loc[(stats_salary_fa['Season'] >= 2003), 'Era'] = 'afmb'
```
